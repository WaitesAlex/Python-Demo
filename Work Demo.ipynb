```python
#Importing the 'modules' I need
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#This is how we read in the data. Notice how we have to specify the function from pandas
df = pd.read_excel('AnalystAuthorTest_SuicideData Oct 2018.xls', sheet_name='Data')
```

```python
#In R this would be head(df) which is already a bit different
df.head()
```

```python
#This method throws an error
head(df)
```

```python
df.groupby(by = ['RegYear','MethodOfSuicide']).count()
```

```python
df.groupby(by = ['RegYear','MethodOfSuicide']).count().unstack()
```

```python
df = df.groupby(by = ['RegYear','MethodOfSuicide']).count().unstack()
```

```python
df = df['CauseOfDeath']
```

```python
list(df.columns.values)
```

```python
cols = ['Hanging, strangulation and suffocation',
        'Poisoning – gases and vapours',
        'Poisoning – solids and liquids',
        'Firearms and explosives ',
        'Submersion (drowning) ',
        'Other means']
df = df[cols]
df
```

```python
dash_line = ['', (2,2), (2,2,4,4), (4,2), (4,4,2,4), (2,4,6,4)]
```

```python
color_pal = ["#9b59b6","#2ecc71", "#3498db", "#95a5a6", "#e74c3c", "#34495e"]
```

```python
fig = plt.figure(figsize = (14, 8))

sns.set_context(font_scale = 3)


sns.lineplot(data=df, palette = color_pal, linewidth = 2,
             dashes= dash_line)

sns.despine()

plt.ylabel("Suicides")
plt.xlabel("Year")
plt.title("Suicides in New Zealand by Method and Year")
plt.legend(frameon = False)
```

```python
fig = plt.figure(figsize = (14, 8))

sns.set_context(font_scale = 3)
sns.lineplot(data=df.drop(['Hanging, strangulation and suffocation'], axis = 1), 
             palette = color_pal[1:], linewidth = 2,
             dashes = dash_line[1:])
sns.despine()

plt.ylabel("Suicides")
plt.title("Suicides in New Zealand by Method and Year Excluding Hanging, Strangulation and Suffocation")
plt.legend(frameon = False)
```
